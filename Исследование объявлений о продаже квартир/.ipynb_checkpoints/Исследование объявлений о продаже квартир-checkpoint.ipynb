Заполнение пропусков. 



```python
print('mediana_kitchen_area:')
for rooms in sorted(data['rooms'].unique()):#заполняю медтианами (по комнатам) площадь кухонь, где площади кухонь и жилые - пропущены
    mediana_kitchen_area = data.loc[data['rooms'].eq(rooms),'kitchen_area'].median() 
    data.loc[data['rooms'].eq(rooms) &  data['living_area'].isna() & data['kitchen_area'].isna(), 'kitchen_area'] = mediana_kitchen_area 

for rooms in sorted(data['rooms'].unique()):#заполняю медтианами (по комнатам) площадь кухонь, где площади кухонь - пропущены, аплощажи жилые - не пропущены
    mediana_kitchen_area = data.loc[data['rooms'].eq(rooms),'kitchen_area'].median() 
    print(mediana_kitchen_area)
    data.loc[data['rooms'].eq(rooms) & (data['total_area'] > data['living_area']+mediana_kitchen_area) & data['living_area'].notna() & data['kitchen_area'].isna(), 'kitchen_area'] = mediana_kitchen_area 


for rooms in sorted(data['rooms'].unique()):#заполняю медтианами (по комнатам) жилую площадь, где площади кухонь - не пропущены, а площажи жилые - пропущены
    mediana_living_area = data.loc[data['rooms'].eq(rooms),'living_area'].median() 
    print(rooms, mediana_living_area)
    data.loc[data['rooms'].eq(rooms) & (data['total_area'] > (data['kitchen_area']+mediana_living_area)) & data['living_area'].isna() & data['kitchen_area'].notna(), 'living_area'] = mediana_living_area            
```

## Добавление в таблицу новых столбцов

Добавлю новый столбец - цена 1 вк м. Для этого разделю столбец "last_price" на "total_area", округлила результат до целых

```python
data['price_1sqm'] = round(data['last_price']/data['total_area'])
```

Добавлю новый столбец, который отображает день недели публикации объявления (0 — понедельник, 1 — вторник и так далее)

```python
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
```

Добавлю новыe столбцы, которые отображают месяц/год публикации объявления 

```python
data['year_exposition'] = data['first_day_exposition'].dt.year
data['month_exposition'] = data['first_day_exposition'].dt.month
data.head()#проверим результат
```

Добавлю новый столбец, который отображает тип этажа квартиры (значения — «первый», «последний», «другой»)

```python
def floor_type(data):
    try:
        if data['floor'] == 1:
            return 'первый'
        elif data['floor'] == data['floors_total']:
            return 'последний'
        return 'другой'
    except:
        pass
```

```python
data['floor_type'] = data.apply(floor_type, axis=1)
data[['floor','floors_total','floor_type']].head(10)#проверим результат
```

Добавлю новый столбец  "cityCenters_nearest_km" расстояние до центра города в километрах (переведу из м в км и округлите до целых значений).

```python
data['cityCenters_nearest_km'] = (data['cityCenters_nearest']/1000).round()
display(data.head())#проверим выполнени
```

### Изучение факторов, влияющих на общую (полную) стоимость объекта


Для этого сделаю сводные таблицы
цена от:
- общей площади; last_price_total_area
- жилой площади; last_price_living_area
- площади кухни; last_price_kitchen_area
- количества комнат;  last_price_rooms
- этажа, на котором расположена квартира (первый, последний, другой); last_price_floor_type
- даты размещения (день недели, месяц, год). last_price_days_exposition/last_price_month_exposition/last_price_year_exposition

Прежде чем строить зависимости, отброшу значения, не входящие в пределы нормальных значений (1,5 межквартильного размаха).
Для этого использую функцию для подсчета 1,5 квартильного размаха

Выбросы буду исключать из столбцов:
    - 'last_price',
    - 'total_area', 
    - 'kitchen_area', 
    - 'living_area', 
    - 'days_exposition', 
    - 'rooms'

```python
def quartile_range(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq

columns_blowout = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 'rooms', 'airports_nearest', 'cityCenters_nearest', 'parks_nearest','days_exposition']

for col in columns_blowout:
    data[col] = quartile_range(data, col)

data['rooms'] = data['rooms'].fillna(0).astype(int) #переведем в int число комнат

data.describe()
```
